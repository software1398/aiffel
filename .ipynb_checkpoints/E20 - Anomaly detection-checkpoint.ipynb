* 보간으로 대체

### Novelties(특이치)
일반적으로 다변량 분석의 결과로 찾아낼 수 있고, 이전까지 보지 못했던 패턴이나 데이터   
특이치는 데이터를 수정하지 않는다.

outlier와 novelies   
https://hoya012.github.io/blog/anomaly-detection-overview-1/

데이터 내에서 이상치를 찾으려면 분포를 확인하여 신뢰구간을 벗어난 데이터를 이상치로 볼 수 있다.   
1. 정규분포를 따르는지 확인하는 방법 : z-test
2. 시계열 데이터 중에서 정규분포에 가까운 데이터를 뽑아내는 방법 : Time series decomposition

```python
%matplotlib inline
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-5, 5, 0.001)
y = norm.pdf(x,0,1) 
# 평균이 0이고, 표준편차가 1인 정규분포를 그립니다.

# build the plot
fig, ax = plt.subplots(figsize=(9,6))
ax.fill_between(x,y,0, alpha=0.3, color='b')
ax.set_xlim([-4,4])
ax.set_title('normal distribution')
plt.show()
```

```python
# 종가의 데이터 분포
fig, ax = plt.subplots(figsize=(9,6))
_ = plt.hist(df.Close, 100, density=True, alpha=0.75)
```

```python
# 단변수의 정규분포 여부는 z-test로 확인 가능

from statsmodels.stats.weightstats import ztest
_, p = ztest(df.Close)
print(p)
```

p가 0.05이하로 나오면 normal distribution과 거리가 멀다는 뜻  
'정규 분포'로 가정한 통계적 추정이 어려움, 신뢰구간 분석 적용이 안됨

시계열 데이터 중에서 정규 분포에 가까운 데이터 뽑아내는 방법 - Time series decomposition  
이 방법으로 시계열 데이터를 3가지 성분으로 분리되고 그 중 residual 데이터는 stationary에 가까운 형태를 가짐  
stationary에 가까운 데이터는 정규 분포를 가진다.  
https://rfriend.tistory.com/510   
trend, seasonality, residual 세 가지 성분으로 분해
time series decomposition에는 additive와 multipllicative 방법이 있다.

```python
# 계절성의 주기는 50으로 가정
# pip install statsmodels
from statsmodels.tsa.seasonal import seasonal_decompose
result = seasonal_decompose(df.Close, model='additive', two_sided=True, 
                            period=50, extrapolate_trend='freq') # 계절적 성분 50일로 가정
result.plot()
plt.show()
```

```python
#그래프가 너무 작아서 안보인다면

fig, axes = plt.subplots(ncols=1, nrows=4, sharex=True, figsize=(12,9))
result.observed.plot(ax=axes[0], legend=False)
axes[0].set_ylabel('Observed', fontsize=15)
result.trend.plot(ax=axes[1], legend=False)
axes[1].set_ylabel('Trend', fontsize=15)
result.seasonal.plot(ax=axes[2], legend=False)
axes[2].set_ylabel('Seasonal', fontsize=15)
result.resid.plot(ax=axes[3], legend=False)
axes[3].set_ylabel('Residual', fontsize=15)
plt.show()
```

```python
# seasonal 성분은 너무 빼곡하게 보이네요. 다시 확인해보겠습니다.
result.seasonal[:100].plot()
#-8 에서 10 사이를 주기적으로 반복하는게 보이네요.
```

```python
# 잔차는 평균 0을 기준으로 분포하고 있다.
# 정규분포가 가능하므로 신뢰구간 분석이 가능하다.

fig, ax = plt.subplots(figsize=(9,6))
_ = plt.hist(result.resid, 100, density=True, alpha=0.75)
```

```python
r = result.resid.values
st, p = ztest(r)
print(st,p)
```

```python
# 평균을 기준으로 3σ를 벗어나는 데이터 찾기

# 평균과 표준편차 출력
mu, std = result.resid.mean(), result.resid.std()
print("평균:", mu, "표준편차:", std)

# 3-sigma(표준편차)를 기준으로 이상치 판단
print("이상치 갯수:", len(result.resid[(result.resid>mu+3*std)|(result.resid<mu-3*std)]))
```

```python
df.Date[result.resid[
    (result.resid>mu+3*std)|(result.resid<mu-3*std)].index]
```

Multi-variable Anomaly Detection(다변량 이상치 탐색)  
* Clustering: 정상적인 데이터끼리 이상한 데이터끼리 그룹핑된다.
* Forecasting: 시계열 예측모델을 만들어서 예측오차가 크게 발생하는 지점을 찾는다, Auto-encoder로 탐색

통계 방식과 동일하게 trend/seasonality를 제거해야 이상한 데이터를 찾아낼 수 있다.

```python
# 데이터 전처리
def my_decompose(df, features, freq=50):
    trend = pd.DataFrame()
    seasonal = pd.DataFrame()
    resid = pd.DataFrame()
    
    # 사용할 feature 마다 decompose 를 수행합니다.
    for f in features:
        result = seasonal_decompose(df[f], 
                                    model='additive', period=freq, extrapolate_trend=freq)
        trend[f] = result.trend.values
        seasonal[f] = result.seasonal.values
        resid[f] = result.resid.values
        
    return trend, seasonal, resid

# 각 변수별 트렌드/계절적/잔차
tdf, sdf, rdf = my_decompose(df, features=['Open','High','Low','Close','Volume'])
tdf.describe()
```

```python
rdf.describe()
```

```python
# volume의 데이터 크기가 너무 크다. volumne에 영향을 많이 받게 된다.
# 각 변수들을 정규화한다.

# 표준정규화
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(rdf)
print(scaler.mean_)
norm_rdf = scaler.transform(rdf)
norm_rdf
```

딥러닝 이상치 탐색의 단점

2. 학습에 활용할 만큼 충분한 데이터가 필요

### Auto-Encoder 이용하기

* 오토인코더는 인코더와 디코더로 구성
* 인코더는 입력데이터로부터 중요한 정보를 추출한다. 압축된 데이터을 확보
* 디코더는 압축된 데이터를 가지고 입력데이터와 유사한 형태의 데이터를 재생성

오토인코더의 핵심은 입력데이터로부터 추출한 중요한 정보로부터 이와 유사한 데이터를  
재생성하기 때문에 비정상적인 데이터는 자연적으로 제거된다.

```python

# 필요한 라이브러리를 불러옵니다

import tensorflow as tf
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, LSTM, RepeatVector, TimeDistributed
from tensorflow.keras.losses import Huber
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
```

```python
# 모델 재현성을 위해 난수 시드설정을 해줍니다
tf.random.set_seed(777)
np.random.seed(777)
```

### LSTM을 이용하여 오토인코더 모델 만들기

```python
# 입력 변수를 정규화시키면 딥러닝 모델 학습이 원활하다

from sklearn.preprocessing import StandardScaler

# 데이터 전처리 - 하이퍼 파라미터
window_size = 10
batch_size = 32
features = ['Open','High','Low','Close','Volume']
n_features = len(features)
TRAIN_SIZE = int(len(df)*0.7)

# 데이터 전처리
# 표준정규분포화합니다.

scaler = StandardScaler()
scaler = scaler.fit(df.loc[:TRAIN_SIZE,features].values)
scaled = scaler.transform(df[features].values)
```

keras timeseriesGenerator를 이용하여 간단하게 데이터셋을 만든다
1. data: 입력데이터, 'open', 'high', 'low', 'close', 'volume'
2. targets: 출력데이터, 인풋을 그대로 예측하기 때문에 data와 동일하다
3. length: 몇 개의 time_step
4. stride: time_step사이즈 결정

```python
# keras TimeseriesGenerator 를 이용해서 간편하게 데이터 셋을 만듭니다

train_gen = TimeseriesGenerator(
    data = scaled,
    targets = scaled,
    length = window_size,
    stride=1,
    sampling_rate=1,
    batch_size= batch_size,
    shuffle=False,
    start_index=0,
    end_index=None,
)

valid_gen = TimeseriesGenerator(
    data = scaled,
    targets = scaled,
    length = window_size,
    stride=1,
    sampling_rate=1,
    batch_size=batch_size,
    shuffle=False,
    start_index=TRAIN_SIZE,
    end_index=None,
)

print(train_gen[0][0].shape)
print(train_gen[0][1].shape)
```

```python
# 모델 만들기
# 2개 층의 LSTM으로 인코더 만든다
# Repeatvector는 input을 window_size만큼 복사해준다.

model = Sequential([
    # >> 인코더 시작
    LSTM(64, activation='relu', return_sequences=True, 
         input_shape=(window_size, n_features)),
    LSTM(16, activation='relu', return_sequences=False),
    ## << 인코더 끝
    ## >> Bottleneck
    RepeatVector(window_size),
    ## << Bottleneck
    ## >> 디코더 시작
    LSTM(16, activation='relu', return_sequences=True),
    LSTM(64, activation='relu', return_sequences=False),
    Dense(n_features)
    ## << 디코더 끝
])

model.summary()
```

```python
# 체크포인트
# 학습을 진행하며 validation 결과가 가장 좋은 모델을 저장해둠
import os

checkpoint_path = os.getenv('HOME')+'/aiffel/anomaly_detection/kospi/mymodel.ckpt'
checkpoint = ModelCheckpoint(checkpoint_path, 
                             save_weights_only=True, 
                             save_best_only=True, 
                             monitor='val_loss',
                             verbose=1)

# 얼리스탑
# 학습을 진행하며 validation 결과가 나빠지면 스톱. patience 횟수만큼은 참고 지켜본다

early_stop = EarlyStopping(monitor='val_loss', patience=5)

model.compile(loss='mae', optimizer='adam',metrics=["mae"])
```

```python
hist = model.fit(train_gen, 
          validation_data=valid_gen,
          steps_per_epoch=len(train_gen),
          validation_steps=len(valid_gen),
          epochs=50, 
          callbacks=[checkpoint, early_stop])
```

```python
model.load_weights(checkpoint_path)
```

```python
fig = plt.figure(figsize=(12,8))
plt.plot(hist.history['loss'], label='Training')
plt.plot(hist.history['val_loss'], label='Validation')
plt.legend()
```

시계열 데이터를 window_size만큼 밀어가면서 예측하는 모델을 만들었다.  
그래서 train_gen길이는 원본 df길이보다 window_size만큼 짧다.  
예측결과와 비교시 scaled의 앞에서 window_size만큼 건너 뛰어야 한다.

```python
# 예측 결과를 pred 로, 실제 데이터를 real로 받습니다
pred = model.predict(train_gen)
real = scaled[window_size:]

mae_loss = np.mean(np.abs(pred-real), axis=1)

# 샘플 개수가 많기 때문에 y축을 로그 스케일로 그립니다
fig, ax = plt.subplots(figsize=(9,6))
_ = plt.hist(mae_loss, 100, density=True, alpha=0.75, log=True)
```

```python
# 무리와 동떨어진 데이터가 보인다.
# mae_loss의 기준치를 3으로 해서 분석한다.
# 스케일링 되지 않은 원본 데이터와 비교

import copy

# window_size만큼 제외하고 loss컬럼을 추가한다.
test_df = copy.deepcopy(df.loc[window_size:]).reset_index(drop=True)
test_df['Loss'] = mae_loss

threshold = 3
test_df.loc[test_df.Loss>threshold]
```

```python
# open, close, low, high는 스케일이 비슷하므로 같이 그리고
# volume, loss는 각각
# 이상치는 붉은점으로 표시

fig = plt.figure(figsize=(12,15))

# 가격들 그래프입니다
ax = fig.add_subplot(311)
ax.set_title('Open/Close')
plt.plot(test_df.Date, test_df.Close, linewidth=0.5, alpha=0.75, label='Close')
plt.plot(test_df.Date, test_df.Open, linewidth=0.5, alpha=0.75, label='Open')
plt.plot(test_df.Date, test_df.Close, 'or', markevery=[mae_loss>threshold])

# 거래량 그래프입니다
ax = fig.add_subplot(312)
ax.set_title('Volume')
plt.plot(test_df.Date, test_df.Volume, linewidth=0.5, alpha=0.75, label='Volume')
plt.plot(test_df.Date, test_df.Volume, 'or', markevery=[mae_loss>threshold])

# 오차율 그래프입니다
ax = fig.add_subplot(313)
ax.set_title('Loss')
plt.plot(test_df.Date, test_df.Loss, linewidth=0.5, alpha=0.75, label='Loss')
plt.plot(test_df.Date, test_df.Loss, 'or', markevery=[mae_loss>threshold])
```
