```python
# 본인의 취향과 유사한 아티스트를 추천받을려면 사용자 취향과 비슷한 아티스트 정보를
# 5개 이상 입력받아야 한다.
# 본인이 좋아하시는 아티스트 데이터로 바꿔서 추가하셔도 됩니다! 단, 이름은 꼭 데이터셋에 있는 것과 동일하게 맞춰주세요. 
my_favorite = ['black eyed peas' , 'maroon5' ,'jason mraz' ,'coldplay' ,'beyoncé']

# 'zimin'이라는 user_id가 위 아티스트의 노래를 30회씩 들었다고 가정하겠습니다.
my_playlist = pd.DataFrame({'user_id': ['zimin']*5, 'artist': my_favorite, 'play':[30]*5})

if not data.isin({'user_id':['zimin']})['user_id'].any():  # user_id에 'zimin'이라는 데이터가 없다면
    data = data.append(my_playlist)                        # 위에 임의로 만든 my_favorite 데이터를 추가해 줍니다. 

data.tail(10)       # 잘 추가되었는지 확인해 봅시다.
```

```python
# # 고유한 유저, 아티스트를 찾아내는 코드
user_unique = data['user_id'].unique()
artist_unique = data['artist'].unique()

# 유저, 아티스트 indexing 하는 코드 idx는 index의 약자입니다.
user_to_idx = {v:k for k,v in enumerate(user_unique)}
artist_to_idx = {v:k for k,v in enumerate(artist_unique)}
```

```python
# 인덱싱이 잘 되었는지 확인해 봅니다. 
print(user_to_idx['zimin'])    # 358869명의 유저 중 마지막으로 추가된 유저이니 358868이 나와야 합니다. 
print(artist_to_idx['black eyed peas'])
```

```python
# indexing을 통해 데이터 컬럼 내 값을 바꾸는 코드
# dictionary 자료형의 get 함수는 https://wikidocs.net/16 을 참고하세요.

# user_to_idx.get을 통해 user_id 컬럼의 모든 값을 인덱싱한 Series를 구해 봅시다. 
# 혹시 정상적으로 인덱싱되지 않은 row가 있다면 인덱스가 NaN이 될 테니 dropna()로 제거합니다. 
temp_user_data = data['user_id'].map(user_to_idx.get).dropna()
if len(temp_user_data) == len(data):   # 모든 row가 정상적으로 인덱싱되었다면
    print('user_id column indexing OK!!')
    data['user_id'] = temp_user_data   # data['user_id']을 인덱싱된 Series로 교체해 줍니다. 
else:
    print('user_id column indexing Fail!!')

# artist_to_idx을 통해 artist 컬럼도 동일한 방식으로 인덱싱해 줍니다. 
temp_artist_data = data['artist'].map(artist_to_idx.get).dropna()
if len(temp_artist_data) == len(data):
    print('artist column indexing OK!!')
    data['artist'] = temp_artist_data
else:
    print('artist column indexing Fail!!')

data
```

```python
# 암묵적/명시적 데이터 분석
# https://orill.tistory.com/entry/Explicit-vs-Implicit-Feedback-Datasets?category=1066301
# 1회만 play한 데이터의 비율을 보는 코드
only_one = data[data['play']<2]
one, all_data = len(only_one), len(data)
print(f'{one},{all_data}')
print(f'Ratio of only_one over all data is {one/all_data:.2%}')  # f-format에 대한 설명은 https://bit.ly/2DTLqYU
# 암묵적 데이터 처리 방향에 대해 결정한다. 선호도, 가중치 등
```

```python
# 추천시스템 모델 중 Matrix Factorization(MF, 행렬분해)
# R(평가 행렬) = P(유저 특성), Q(아이템 특성)분리
# m, n -> m, k 와 n, k

# 평가행렬에서 0이 아닌 유효한 값과 좌표만으로 이루어진 매트릭스 생성 필요
# CRS(Compressed Sparse Row) Matrix
# https://lovit.github.io/nlp/machine%20learning/2018/04/09/sparse_mtarix_handling/#csr-matrix
# 실습 위에 설명보고 이해해서 만들어보기
from scipy.sparse import csr_matrix

num_user = data['user_id'].nunique()
num_artist = data['artist'].nunique()

csr_data = csr_matrix((data.play, (data.user_id, data.artist)), shape= (num_user, num_artist))
csr_data
```

```python
# MF모델을 implicit패키지를 이용하여 학습하기
# implicit패키지 내 als(AlternatingLeastSquares)모델 사용
# als : MF에서 쪼개진 두 Matrix를 한쪽은 고정시키고 다른 쪽을 학습하는 방식을 번갈아 수행

from implicit.als import AlternatingLeastSquares
import os
import numpy as np

# implicit 라이브러리에서 권장하고 있는 부분입니다. 학습 내용과는 무관합니다.
os.environ['OPENBLAS_NUM_THREADS']='1'
os.environ['KMP_DUPLICATE_LIB_OK']='True'
os.environ['MKL_NUM_THREADS']='1'
```

```python
# als의 __init__ 파라미터의 의미
# factors : 유저와 아이템의 벡터를 몇 차원으로 할 것인가
# regularization : 과적합을 방지하기 위해 정규화 값을 얼마나 사용할 것인가
# use_gpu : GPU를 사용할 것인가
# iterations : epochs값을 얼마나 할 것인가
# factors와 iterations를 늘리면 학습이 잘 되지만, 과적합 주의

# Implicit AlternatingLeastSquares 모델의 선언
als_model = AlternatingLeastSquares(factors=100, regularization=0.01, use_gpu=False, iterations=15, dtype=np.float32)

# als 모델은 input으로 (item X user 꼴의 matrix를 받기 때문에 Transpose해줍니다.)
csr_data_transpose = csr_data.T
csr_data_transpose

# 모델 훈련
als_model.fit(csr_data_transpose)
```

```python
# 모델에서 나의 벡터와 black eyed peas의 벡터를 어떻게 만들었는지
# 두 벡터를 곱하면 어떤 값이 나오는지 살펴보자

zimin, black_eyed_peas = user_to_idx['zimin'], artist_to_idx['black eyed peas']
zimin_vector, black_eyed_peas_vector = als_model.user_factors[zimin], als_model.item_factors[black_eyed_peas]
```

```python
zimin_vector
```

```python
black_eyed_peas_vector
```

```python
# zimin과 black_eyed_peas를 내적하는 코드
np.dot(zimin_vector, black_eyed_peas_vector)

# 1이 안된다.
```

```python
queen = artist_to_idx['queen']
queen_vector = als_model.item_factors[queen]
np.dot(zimin_vector, queen_vector)

# 이 확률이 얼마 정도돼야 선호한다고 볼 수 있나?
# 객관적인 기준은 없지만 데이터에 따라 다르다고 볼 수 있다.
```

```python
# als내에 similar_items메서드를 이용하여 비슷한 아티스트를 찾아보자.

favorite_artist = 'coldplay'
artist_id = artist_to_idx[favorite_artist]
similar_artist = als_model.similar_items(artist_id, N=15)
similar_artist
# (아티스트id, 유사도) tuple로 반환
```

```python
#artist_to_idx 를 뒤집어, index로부터 artist 이름을 얻는 dict를 생성합니다. 
idx_to_artist = {v:k for k,v in artist_to_idx.items()}
[idx_to_artist[i[0]] for i in similar_artist]
```

```python
def get_similar_artist(artist_name: str):
    artist_id = artist_to_idx[artist_name]
    similar_artist = als_model.similar_items(artist_id)
    similar_artist = [idx_to_artist[i[0]] for i in similar_artist]
    return similar_artist

get_similar_artist('2pac')
```

```python
get_similar_artist('lady gaga')
```

```python
# als내 recommend메서드를 사용하여 내가 좋아할 만한 아티스트 추천해보자.
# filter_already_liked_items는 유저가 이미 평가한 아이템은 제외한다.

user = user_to_idx['zimin']
# recommend에서는 user*item CSR Matrix를 받습니다.
artist_recommended = als_model.recommend(user, csr_data, N=20, filter_already_liked_items=True)
artist_recommended
```

```python
[idx_to_artist[i[0]] for i in artist_recommended]
```

```python
# 추천 아티스트 중에 rihanna을 추천하게 한 나의 데이터의 기여 정도 확인해보자
# als내 explain메서드

rihanna = artist_to_idx['rihanna']
explain = als_model.explain(user, csr_data, itemid=rihanna)

# 추천 콘텐츠 점수 = 기여한 다른 콘텐츠 + 기여도
[(idx_to_artist[i[0]], i[1]) for i in explain[1]]
```
