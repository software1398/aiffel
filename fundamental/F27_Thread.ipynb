```python
# ìŠ¤ë ˆë“œ, í”„ë¡œì„¸ìŠ¤ í’€ì„ ë§Œë“¤ì–´ì„œ ìë™ìœ¼ë¡œ í• ë‹¹í•˜ê³  ì¢…ë£Œí•˜ê²Œ í•œë‹¤.
# 1. Queueë¼ëŠ” ìë£Œêµ¬ì¡°ë¥¼ ì‚¬ìš©í•˜ì—¬ í’€ì„ ë§Œë“œëŠ” ë°©ë²•
# 2. concurrent.futuresë¼ì´ë¸ŒëŸ¬ë¦¬ ThreadPoolExecutor, ProcessPoolExecutor í´ë˜ìŠ¤ ì´ìš©í•˜ëŠ” ë°©ë²•

from concurrent.futures import ThreadPoolExecutor

class Delivery:
    def run(self):
        print("delivering")
w = Delivery()

with ThreadPoolExecutor() as executor:
    future = executor.submit(w.run)
```

```python
# multiprocessing.Pool.map: ì—¬ëŸ¬ ê°œì˜ í”„ë¡œì„¸ìŠ¤ì— íŠ¹ì • í•¨ìˆ˜ë¥¼ ë§¤í•‘í•˜ì—¬ ë³‘ë ¬ ì²˜ë¦¬

from multiprocessing import Pool
from os import getpid

def double(i):
    print("I'm process ", getpid())    # pool ì•ˆì—ì„œ ì´ ë©”ì†Œë“œê°€ ì‹¤í–‰ë  ë•Œ pidë¥¼ í™•ì¸í•´ ë´…ì‹œë‹¤.
    return i * 2

with Pool() as pool:
      result = pool.map(double, [1, 2, 3, 4, 5])
      print(result)
```
