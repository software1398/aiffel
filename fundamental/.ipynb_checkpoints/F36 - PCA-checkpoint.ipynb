## F36 - PCA 

###### 고유값(eigenvalue)과 고유벡터(eigenvector)

다음의 수식을 만족하는
<math xmlns="http://www.w3.org/1998/Math/MathML">
  <mi>&#x3BB;</mi>를 고유값, 
  <mi mathvariant="bold">x</mi>를 고유벡터라고 한다.
</math>

<math xmlns="http://www.w3.org/1998/Math/MathML" display="block">
  <mrow>
    <mi mathvariant="bold">A</mi>
    <mi mathvariant="bold">x</mi>
  </mrow>
  <mo>=</mo>
  <mi>&#x3BB;</mi>
  <mi mathvariant="bold">x</mi>
  
</math>

이런 고유값과 고유벡터는 여러 개를 가질 수 있는데 간단하게 행렬을 이용한 matrix form으로 나타낼 수 있다.   
이를 고유값 분해(Eigenvalue Decomposition, EVD)라고 한다.

주성분 분석: https://ratsgo.github.io/machine%20learning/2017/04/24/PCA/   
가장 큰 고유값에 해당하는 고유벡터가 데이터를 가장 잘 표현해 낼 수 있는 축이다.   
데이터를 이용해 공분산 행렬을 만든다.

```python
# 영상 잡음 제거
# 비교적 큰 고유값에 해당하는 기저벡터만으로 데이터를 복원하면 잡음이 제거된 손글씨 이미지를 얻을 수 있다.

from sklearn.decomposition import PCA
from sklearn.datasets import load_digits
from matplotlib import pyplot as plt
import numpy as np

digits = load_digits()
print("손글씨 이미지의 크기는",digits.data.shape[1],"픽셀 입니다.")
```

```python
import matplotlib.pyplot as plt

def plot_digits(data):
    fig, axes = plt.subplots(4, 10, figsize=(10, 4),
                             subplot_kw={'xticks':[], 'yticks':[]},
                             gridspec_kw=dict(hspace=0.1, wspace=0.1))
    for i, ax in enumerate(axes.flat):
        ax.imshow(data[i].reshape(8, 8),
                  cmap='binary', interpolation='nearest',
                  clim=(0, 16))
plot_digits(digits.data) # 원본 손글씨 이미지 출력
```

```python
np.random.seed(42)
noisy = np.random.normal(digits.data, 4)
plot_digits(noisy) # 잡음이 섞인 손글씨 이미지 출력
```

```python
pca = PCA(0.50).fit(noisy) # 데이터의 50% 수준(분산 기준)를 표현할 수 있는 PCA 기저벡터의 수는?
print("데이터의 50% 수준(분산 기준)까지 복원 가능한 PCA 기저벡터의 수는",pca.n_components_,"개 입니다")
```

```python
components = pca.transform(noisy)
filtered = pca.inverse_transform(components)
plot_digits(filtered) # 복원된 손글씨 이미지
```

```python
# 영상 압축
# 64pixel 값을 모두 저장하지 않고 비교적 큰 고유값 N개만 지정하여 압축하고, N개의 고유벡터의 선형조합으로 압축 해제 가능?

from sklearn.decomposition import PCA
from sklearn.datasets import load_digits

digits = load_digits()
print("손글시 이미지의 크기는",digits.data.shape[1],"픽셀 입니다.")
```

```python
def plot_digits(data):
    fig, axes = plt.subplots(4, 10, figsize=(10, 4),
                             subplot_kw={'xticks':[], 'yticks':[]},
                             gridspec_kw=dict(hspace=0.1, wspace=0.1))
    for i, ax in enumerate(axes.flat):
        ax.imshow(data[i].reshape(8, 8),
                  cmap='binary', interpolation='nearest',
                  clim=(0, 16))
plot_digits(digits.data) # 원본 이미지 출력
```

```python
pca = PCA(0.90).fit(digits.data) #데이터의 90% 수준까지 복원 가능한 PCA 기저벡터의 수는?
print("데이터의 90% 수준까지 복원 가능한 PCA 기저벡터의 수는",pca.n_components_,"개 입니다")
```

```python
compression_ratio=(digits.data.shape[1]-pca.n_components_)/digits.data.shape[1]*100
print("압축률은",compression_ratio,"% 입니다.")

# (64-N)/64 * 100
```

```python
components = pca.transform(digits.data)
filtered = pca.inverse_transform(components)
plot_digits(filtered) #복원된 이미지
```
